Step 1 :  Import the necessary libraries

```python
import pandas as pd
import datetime
```

Step 2 : Import the dataset 

```python
filename = 'wind.data'
```

Step 3 : Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data = pd.read_csv(filename,sep='\s+',parse_dates=[[0,1,2]])
data.head()
```

Step 4 : Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
import datetime
def fix_date(x):
    if x.year > 1989:
        year = x.year - 100
    else:
        year = x.year
    return datetime.date(year,x.month,x.day)
data['Date/Time'] = data['Yr_Mo_Dy'].apply(fix_date)
data=data.drop(['Yr_Mo_Dy'],axis=1)
data.head()
```

Step 5 : Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
data = data.set_index('Date/Time')
data.head()
```

Step 6 : Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.

```python
data.isnull().sum()
```

Step 7 : Compute how many non-missing values there are in total.

```python
data.notnull().sum()
data.shape[0] - data.isnull().sum()
```

Step 8 : Calculate the mean windspeeds of the windspeeds over all the locations and

```python
data.mean(axis=0)
```

Step 9 : Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days

```python
loc_stats=data.describe(percentiles=[])
loc_stats.head()
```

Step 10 : Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day. A different set of numbers for each day.

```python
day_stats=pd.DataFrame()
day_stats['mean']=data.mean(axis=1)
day_stats['min']=data.min(axis=1)
day_stats['max']=data.max(axis=1)
day_stats['standard deviation']=data.std(axis=1)
day_stats.head()
```

Step 11 : Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January.

```python
data.loc[data.index.month == 1].mean()
```

Step 14 : Downsample the record to a weekly frequency for each location.

```python
data.groupby(data.index.to_period('W')).mean()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
us_marriage_divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
us_marriage_divorce_data=us_marriage_divorce_data[(us_marriage_divorce_data['Year']>=1867) & (us_marriage_divorce_data['Year']<=2014)]
fig, ax = plt.subplots()
plt.xlabel('Years')
plt.ylabel('Population per capita')
ax.plot(us_marriage_divorce_data['Year'], us_marriage_divorce_data['Marriages_per_1000'])
ax.plot(us_marriage_divorce_data['Year'],us_marriage_divorce_data['Divorces_per_1000'])
ax.set_title('Number of Marriages and Divorces per capita in the U.S. between 1867 and 2014')
ax.legend()
```

# Question 7

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

us_marriage_divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

labels = [1900,1950,2000]
us_marriage_divorce_data = us_marriage_divorce_data[us_marriage_divorce_data['Year'].apply(lambda x: x in [1900, 1950, 2000])]
years = us_marriage_divorce_data['Year'].values
marriages_per_capita = us_marriage_divorce_data['Marriages_per_1000'].values
divorces_per_capita = us_marriage_divorce_data['Divorces_per_1000'].values

fig, ax = plt.subplots()
bar_width = 0.35
X = np.arange(len(labels))

p1 = plt.bar(X-bar_width/2, marriages_per_capita, bar_width, color='b',label='marriages per capita')
p2 = plt.bar(X + bar_width/2, divorces_per_capita, bar_width, color='g', label='divorces per capita')

ax.set_xlabel('Years')
ax.set_ylabel('Population per capita')
ax.set_title('Marriages and Divorces per capita for year 1900, 1950 and 2000')

ax.set_xticks(X)
ax.set_xticklabels(labels)

ax.legend()
plt.tight_layout()
plt.show()
```
